```python
%pip install PyPDF2

from PyPDF2 import PdfReader, PdfWriter, PageObject
from PyPDF2.generic import RectangleObject
import io
from reportlab.pdfgen import canvas

def add_margin_to_pdf(input_pdf, output_pdf, page_number):
    # 読み込みと書き込みのPDFオブジェクト
    reader = PdfReader(input_pdf)
    writer = PdfWriter()
    
    margin_top_cm = 2.5
    margin_bottom_cm = 1
    margin_left_cm = 2.5
    margin_right_cm = 1.5
    
    # ポイント単位に変換 (1cm = 28.35ポイント)
    margin_top = margin_top_cm * 28.35
    margin_bottom = margin_bottom_cm * 28.35
    margin_left = margin_left_cm * 28.35
    margin_right = margin_right_cm * 28.35

    for page_num in range(len(reader.pages)):
        page = reader.pages[page_num]
        new_width = float(page.mediabox.width) + margin_left + margin_right
        new_height = float(page.mediabox.height) + margin_top + margin_bottom
        
        # 新しいページを作成
        new_page = PageObject.create_blank_page(width=new_width, height=new_height)
        
        # 元のページを新しいページに描画
        new_page.merge_page(page)
        
        # ページの内容を左上から余白分だけ離して配置
        new_page.add_transformation([1, 0, 0, 1, margin_left, margin_bottom])
        
        # 新しいページのメディアボックスを設定
        new_page.mediabox = RectangleObject([0, 0, new_width, new_height])
        
        # ナンバリングを追加
        packet = io.BytesIO()
        can = canvas.Canvas(packet, pagesize=(new_width, new_height))
        can.setFont("Helvetica", 18)
        can.drawString(new_width / 2, margin_bottom / 2, f"FIG. {page_number + page_num}")
        can.save()
        
        packet.seek(0)
        number_pdf = PdfReader(packet)
        number_page = number_pdf.pages[0]
        new_page.merge_page(number_page)
        
        writer.add_page(new_page)

    with open(output_pdf, 'wb') as output:
        writer.write(output)

# 使用例
input_pdf = "1.rotated.pdf"  # 入力PDFファイル名
output_pdf = "1.margin_4.pdf"  # 出力PDFファイル名
add_margin_to_pdf(input_pdf, output_pdf, 1)  # 余白のサイズ (センチメートル単位)とナンバリングの開始番号
```
