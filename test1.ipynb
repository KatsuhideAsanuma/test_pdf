```python
%pip install PyPDF2

from PyPDF2 import PdfReader, PdfWriter

def rotate_pdf_90_degrees(input_pdf, output_pdf):
    reader = PdfReader(input_pdf)
    writer = PdfWriter()

    for page_num in range(len(reader.pages)):
        page = reader.pages[page_num]
        page.rotate(90)  # 右に90度回転
        writer.add_page(page)

    with open(output_pdf, 'wb') as output:
        writer.write(output)

# 使用例
input_pdf = "1.priv.pdf"  # 入力PDFファイル名
output_pdf = "1.priv_rotated.pdf"  # 出力PDFファイル名
rotate_pdf_90_degrees(input_pdf, output_pdf)
```

```python
%pip install PyPDF2 reportlab

from PyPDF2 import PdfReader, PdfWriter, PageObject
from reportlab.pdfgen import canvas
from reportlab.lib.pagesizes import letter
import io

def add_margin_to_pdf(input_pdf, output_pdf, page_number):
    # 読み込みと書き込みのPDFオブジェクト
    reader = PdfReader(input_pdf)
    writer = PdfWriter()
    
    margin_top_cm = 2.5
    margin_bottom_cm = 1
    margin_left_cm = 2.5
    margin_right_cm = 1.5
    
    # ポイント単位に変換 (1cm = 28.35ポイント)
    margin_top = margin_top_cm * 28.35
    margin_bottom = margin_bottom_cm * 28.35
    margin_left = margin_left_cm * 28.35
    margin_right = margin_right_cm * 28.35

    for page_num in range(len(reader.pages)):
        page = reader.pages[page_num]
        
        # 元のページを右に90度回転
        page.rotate(90)
        
        new_width = float(page.mediabox.height) + margin_left + margin_right
        new_height = float(page.mediabox.width) + margin_top + margin_bottom
        
        # 新しいページを作成
        packet = io.BytesIO()
        can = canvas.Canvas(packet, pagesize=(new_width, new_height))
        
        # 元のページを新しいページに描画
        can.saveState()
        can.translate(margin_left, margin_bottom)
        can.rotate(90)
        can.translate(0, -float(page.mediabox.width))
        can.drawImage(page.extract_text(), 0, 0, width=float(page.mediabox.width), height=float(page.mediabox.height))
        can.restoreState()
        
        # ナンバリングを追加
        can.setFont("Helvetica", 18)
        can.drawString(new_width / 2, margin_bottom / 2, f"FIG. {page_number + page_num}")
        can.save()
        
        packet.seek(0)
        new_pdf = PdfReader(packet)
        new_page = new_pdf.pages[0]
        
        writer.add_page(new_page)

    with open(output_pdf, 'wb') as output:
        writer.write(output)

# 使用例
input_pdf = "1.priv.pdf"  # 入力PDFファイル名
output_pdf = "1.priv_margin.pdf"  # 出力PDFファイル名
add_margin_to_pdf(input_pdf, output_pdf, 1)  # 余白のサイズ (センチメートル単位)とナンバリングの開始番号
```
